Trend

```python

import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
df = pd.read_csv('crimes-in-sydney-lga.csv')

# Group by year and count the number of crimes each year
yearly_crime_trend = df.groupby('incyear').size().reset_index(name='Number of Crimes')

# Plot the combined trend of crime incidents from 2008 to 2016
plt.figure(figsize=(10, 5))
plt.plot(yearly_crime_trend['incyear'], yearly_crime_trend['Number of Crimes'], marker='o')

# Add labels on each data point
for i, row in yearly_crime_trend.iterrows():
    plt.text(row['incyear'], row['Number of Crimes'] + 50, int(row['Number of Crimes']), ha='center', va='bottom')

plt.title('Trend of Crime Incidents per Year (2008-2016)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()

```

Number of incidents by accient type

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('crimes-in-sydney-lga.csv')

accident_type_trend = df.groupby('bcsrcat').size().reset_index(name='Number of Incidents')

plt.figure(figsize=(12, 6))
bars = plt.bar(accident_type_trend['bcsrcat'], accident_type_trend['Number of Incidents'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.title('Number of Incidents by Accident Type')
plt.xlabel('Accident Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)  
plt.grid(True)
plt.show()
```

Yearly crime incidents by category

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('crimes-in-sydney-lga.csv')

crime_yearly_trend = df.groupby(['incyear', 'bcsrcat']).size().unstack().fillna(0)

totals = crime_yearly_trend.sum(axis=1)

ax = crime_yearly_trend.plot(kind='bar', stacked=True, figsize=(12, 6))

for i, total in enumerate(totals):
    ax.text(i, total + 50, int(total), ha='center', va='bottom')

plt.title('Yearly Crime Incidents by Category')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Crime Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

 Number of crime by suburb

```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('crimes-in-sydney-lga.csv')

crime_by_suburb = df.groupby('locsurb').size().reset_index(name='Number of Crimes')

plt.figure(figsize=(12, 6))
bars = plt.bar(crime_by_suburb['locsurb'], crime_by_suburb['Number of Crimes'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, int(yval), ha='center', va='bottom')

plt.title('Number of Crimes by Suburb')
plt.xlabel('Suburb')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)  
plt.grid(True)
plt.show()
```

Crime distribution by gender

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('crimes-in-sydney-lga.csv')

gender_crime_trend = df.groupby(['poisex', 'bcsrcat']).size().unstack().fillna(0)

ax = gender_crime_trend.plot(kind='bar', stacked=True, figsize=(12, 6))

totals = gender_crime_trend.sum(axis=1)
for i, total in enumerate(totals):
    ax.text(i, total + 50, int(total), ha='center', va='bottom')

plt.title('Crime Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

gdf = gpd.read_file('STE_2021_AUST_GDA2020.shp')

penalty_data = pd.read_csv('crimes-in-sydney-lga.csv')

gdf_csv = gpd.GeoDataFrame(penalty_data, geometry=gpd.points_from_xy(penalty_data.bcsrgclng, penalty_data.bcsrgclat))

if 'crime_count' not in gdf_csv.columns:
    crime_counts = gdf_csv.groupby('locsurb').size().reset_index(name='crime_count')
    gdf_csv = gdf_csv.merge(crime_counts, on='locsurb', how='left')

gdf_csv['missing'] = np.where(gdf_csv['crime_count'].isna(), 1, 0)

fig, ax = plt.subplots(1, 1, figsize=(12, 12))

gdf.boundary.plot(ax=ax, linewidth=1, color='black')

ax.set_xlim(150.9, 151.3)
ax.set_ylim(-34.1, -33.7)

cmap = plt.cm.viridis  
gdf_csv[gdf_csv['missing'] == 0].plot(ax=ax, column='crime_count', cmap=cmap, legend=True)

gdf_csv[gdf_csv['missing'] == 1].plot(ax=ax, edgecolor='red', facecolor='none', hatch='///')

plt.title('Number of Crimes in Sydney with Missing Values Highlighted')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import plotly.express as px

gdf = gpd.read_file('STE_2021_AUST_GDA2020.shp')

penalty_data = pd.read_csv('crimes-in-sydney-lga.csv')

gdf_csv = gpd.GeoDataFrame(penalty_data, geometry=gpd.points_from_xy(penalty_data.bcsrgclng, penalty_data.bcsrgclat))

if 'crime_count' not in gdf_csv.columns:
    crime_counts = gdf_csv.groupby('locsurb').size().reset_index(name='crime_count')
    gdf_csv = gdf_csv.merge(crime_counts, on='locsurb', how='left')

gdf_csv['lat'] = gdf_csv.geometry.y
gdf_csv['lon'] = gdf_csv.geometry.x

fig = px.scatter_mapbox(gdf_csv, lat="lat", lon="lon", hover_name="locsurb", color="crime_count",
                        color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10, height=800)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
